#### 1: List the causes of death and percentile in Ohio state in the year 2012

```python
#import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
import seaborn as sns
style.use('ggplot')
```

```python
cause = 'Chronic Disease', 'Unintetional Injuries', 'Alzheimers', 'Infuenza', 'Pneumonia', 'Sepsis'
percentile = [62, 5, 4, 2, 1 , 26]
```

#### 2: Draw a pie chart using the two data points

```python
plt.figure(figsize=(8,8))
explode = [0.05,0,0,0,0,0]
plt.pie(percentile, explode=explode, labels=cause, autopct='%1.1f%%', startangle=0)
plt.axis('equal')
plt.title('Pie Chart Example')
plt.show()
```
