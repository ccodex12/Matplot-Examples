```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
import seaborn as sns
style.use('ggplot')
```

```python
df = pd.read_csv(r"C:\Users\Asus\OneDrive\Dataset\auto_data.csv")
```

```python
df.head()
```

```python
def origin_to_cntry(x):
    if x==1:
        return 'USA'
    elif x==2:
        return 'Europe'
    elif x==3:
        return 'Asia'
    return x
df['origin'] = df['origin'].apply(origin_to_cntry)
```

```python
df.head(30)
```

```python
sns.pairplot(df[['mpg', 'weight', 'origin']], hue='origin', height=4)
plt.show()
```
